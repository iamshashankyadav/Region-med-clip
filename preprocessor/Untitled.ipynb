```python

import random
# Helper to categorize age
def categorize_age(age):
    if pd.isna(age):
        return "unknown age"
    if age < 20: return "young"
    elif age < 40: return "adult"
    elif age < 60: return "middle-aged"
    else: return "elderly"

# Define 20 diverse templates
templates = [
    "Clinical image showing {dx} on the {localization}.",
    "{dx_type}-confirmed {dx} lesion.",
    "{age_group} {sex} patient with {dx}.",
    "Lesion observed on the {localization}.",
    "{dx} detected.",
    "Patient presents with {dx} in the {localization}.",
    "{dx} ({dx_type}) on the {localization}.",
    "{sex} patient showing signs of {dx}.",
    "{dx} lesion in the {localization} of a {age_group} patient.",
    "Medical image of {dx}.",
    "A {dx_type} report confirms {dx} in the {localization}.",
    "Observed {dx} on the {localization} of a {age_group} {sex}.",
    "This {age_group} {sex} shows {dx} on the {localization}.",
    "Dermoscopic image of {dx} in the {localization}.",
    "Case of {dx} confirmed via {dx_type}.",
    "Image reveals {dx} lesion in the {localization}.",
    "Patient's {localization} affected by {dx}.",
    "The {localization} shows signs of {dx} ({dx_type}).",
    "{age_group} {sex} patient with {dx} lesion.",
    "Detected {dx} on the {localization}."
]

# Generate global labels
global_labels = []
for _, row in df.iterrows():
    template = random.choice(templates)
    age_group = categorize_age(row['age'])
    label = template.format(
        dx=row['dx'].lower(),
        dx_type=row['dx_type'].lower(),
        localization=row['localization'].lower(),
        sex=row['sex'].lower(),
        age_group=age_group
    )
    global_labels.append(label)

# Add column to dataframe
df['global_label'] = global_labels

# Save the updated CSV
df.to_csv("lesion_metadata_with_global_labels.csv", index=False)
print("✅ Global labels generated and saved to lesion_metadata_with_global_labels.csv")
```
